### Standardize the data

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
# We have to leave the dummies as they are
# WE CANNOT STANDARDIZE THE DUMMIES
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler

class CustomScaler(BaseEstimator, TransformerMixin):
    
    def __init__(self, columns):
        self.scaler = StandardScaler()
        self.columns = columns
        self.mean_ = None
        self.var_ = None
        
    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns], y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self

    def transform(self, X, y=None, copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        X_not_scaled = X.loc[:,~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled, X_scaled], axis=1)[init_col_order]
```

```python
columns_to_omit = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Education']
```

```python
columns_to_scale = [x for x in unscaled_inputs.columns.values if x not in columns_to_omit]
```

```python
sc = CustomScaler(columns_to_scale)
```

```python
sc.fit(unscaled_inputs)
```

```python
scaled_inputs = sc.transform(unscaled_inputs)
scaled_inputs
```

```python
summary_table = summary_table.sort_index()
summary_table
```

```python
summary_table['Odds ratio'] = np.exp(summary_table.Coefficient)
summary_table
```

```python
summary_table.sort_values('Odds ratio', ascending=False)
# most important to least important
# a weight of 0 implies that no matter the feature value, it has no effect
# for a unit change in the standardized feature, 
    # the odds increase by a multiple equal to the odds ratio(odds ratio =1 -> no change)
# we can observe that daily work load average has the coef around 0
    # therefore its odds ratio is 1 too.
    # it's useless
# we can say that distance to work and weekday are useless too but hold them for now

# if you remove the 0 weights, odds ratios should be almost exactly the same as before you remove 
```
